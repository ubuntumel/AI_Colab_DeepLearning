```python
import numpy as np
import matplotlib.pyplot as plt

from keras import Sequential
from keras.layers import Dense, Activation
from keras.initializers import Zeros, RandomNormal, glorot_normal, glorot_uniform
```

```python
n_input = 784
n_dense = 256
```

```python
b_init= Zeros()
```

```python
w_init = RandomNormal(stddev = 1.0)
w_init = glorot_normal()
w_init = glorot_uniform()
```

```python
model = Sequential()
model.add(Dense(n_dense, input_dim = n_input, kernel_initializer = w_init, bias_initializer = b_init))
model.add(Activation('sigmoid'))
#model.add(Activation('tanh'))
#model.add(Activation('relu'))
```

```python
x = np.random.random((1, n_input))
```

```python
a = model.predict(x)
```

```python
_ = plt.hist(np.transpose(a))
# glorot_normal() sigmoid activation
# Activations are concentrated near 0.5 with few values at 0 or 1. Weighted inputs z are at moderate values,
# most neruons maintain non zero gradients, allowing effective backpropagation.
```

glorot_normal() with tahn activation:
Activations form a 0 concentrated, symmetric distribution. Enabling gradients smoothly and improving convergence.

glorot_normal_tahn.png

glorot_normal() relu activation:
Activations are concentrated near zero with a positive spread. Maintains stable variance, preventing large values, allowing neurons to remain active during training.

glorot_normal_relu.png

glorot_uniform() sigmoid activation:
Activations concentrated around 0.5 with few at 0 or 1. Neurons are not saturated, activations are near the center. Sigmoid ensures meaningful gradients and avoids extreme outputs.

glorot_uniform_sigmoid.png

glorot_uniform() tahn activation:
Activations seems to be symmetrically distributed around zero with most values within the range [-1, 1]. Neurons seems to be mostly in the linear region of the tahn function.

glorot_uniform_tahn.png

glorot_uniform() relu activation:
Most activations are near zero, with a postive tail. Negative inputs are clipped to zero which is expected with ReLU. glorot_uniform() keeps the variance of z under control, preventing large activations. Many neurons have an output of 0, but many of the actives ones produce healthy postive values.
