```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gaussian Error Linear Unit (GELU), GELU weighs inputs by their magnitude.
# Tahn approximation of the GELU function
# 0.5 * x * (1 + tanh(sqrt(2/pi) * (x + 0.044715 * x^3)))
def gelu(x):
    return (0.5 * x) * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * x**3)))

x = np.arange(-5, 5, 0.01)
plt.plot(x, gelu(x))
plt.title('Gaussian Error Linear Unit')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```

```python
# Softplus, A soft approximation of the ReLU function
# f(x) = ln(1 + e^x)
def softplus(x):
  return np.log(1 + np.exp(x))

x = np.arange(-5, 5, 0.01)
plt.plot(x, softplus(x))
plt.title('Softplus')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```

```python
# Exponential Linear Unit, attempts to make the mean activations closer to 0, speeds up learning.
x = np.arange(-5, 5, 0.01)
y = np.where(x > 0, x, np.exp(x) - 1)

plt.plot(x, y)
plt.title('Exponential Linear Unit')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```

```python
# Scaled Exponential Linear Unit
# Alpha and Lambda constants for the self derived neural networks
# f(x) = lambda * x if x > 0 else lambda * alpha * (exp(x) - 1)
alpha = 1.6732632423543772848170429916717
lamb = 1.0507009873554804934193349852946

def selu(x):
  return lamb * np.where(x > 0, x, alpha * (np.exp(x) - 1) )

x = np.arange(-5, 5, 0.1)

plt.plot(x, selu(x))
plt.title('Scaled Exponential Linear Unit')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```

```python
# Leaky Rectified Linear Unit (Leaky ReLU)
# Allows a small, non-zero gradient
# 0.01 standard slope
x = np.arange(-5, 5, 0.01)
y = np.maximum(0, np.where(x > 0, x, 0.01 * x))

plt.plot(x, y)
plt.title('Leaky Rectified Linear Unit')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```

```python
# Parametric Rectified Linear Unit (PReLU)
# Allows a small, non-zero gradient
# Similar to Leaky ReLU, but the slope can be a learned parameter
x = np.arange(-5, 5, 0.01)
y = np.maximum(0, np.where( x > 0, x, 0.1 * x))

plt.plot(x, y)
plt.title('Parametric Rectified Linear Unit')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```

```python
# Sigmoid Linear Unit (SiLU)
# f(x) = x * sigmoid(x)
def sigmoid(x):
  return 1/(1 + np.exp(-x))

def SiLU(x):
  return x * sigmoid(x)

x = np.arange(-5, 5, 0.01)
plt.plot(x, SiLU(x))
plt.title('Sigmoid Linear Unit')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```

```python
# Standard normal distribution formula with mean 0 and variance 1.
def simple_gaussian(x):
  return np.exp(-0.5 * x**2) / np.sqrt(2 * np.pi)

x = np.arange(-5, 5, 0.01)
plt.plot(x, simple_gaussian(x))
plt.title('Simple Gaussian')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()

# Gaussian function allowing custom Mean (mu) and Standard Deviation (sigma).
# f(x) = (1 / (sigma * sqrt(2*pi))) * exp(-0.5 * ((x-mu)/sigma)^2)
def Gaussian(x, mu = 0, sigma = 1):
  return np.exp(-0.5 * ((x - mu) / sigma)**2) / (sigma * np.sqrt(2 * np.pi))

x = np.arange(-5, 5, 0.01)
plt.plot(x, Gaussian(x))
plt.title('Gaussian')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()

```
